```python
# morphological analysis

from camel_tools.morphology.database import MorphologyDB
from camel_tools.morphology.analyzer import Analyzer

# load the default morphological database for Modern Standard Arabic
db = MorphologyDB.builtin_db()

analyzer = Analyzer(db)

analyses = analyzer.analyze('علماء') + analyzer.analyze('السلطنة') + analyzer.analyze('الخلافة') + analyzer.analyze('ارشد') + analyzer.analyze('منسوبة') + analyzer.analyze('مريد') + analyzer.analyze('منصور') + analyzer.analyze('عادل') + analyzer.analyze('الفرح') + analyzer.analyze('العطاء') + analyzer.analyze('الصدقة')

for analysis in analyses:
    print(analysis, '\n')
```

"Generation is the process of inflecting a lemma for a set of morphological features." - via CAMeL Tools Google Colab example

The generator seems to work better if the "al" (alif-lam) is removed from the front of the word, and if the word is diacritized. But even with this, not all words have their diacritizations computed. This may be because of the database being used.


```python
from camel_tools.morphology.database import MorphologyDB
from camel_tools.morphology.generator import Generator

# the database loaded to be used for generation:
db = MorphologyDB.builtin_db(flags='g')

generator = Generator(db)


lemmas = ['صَدَقَة', 'مَنسُوبَة', 'عَطَاء', 'مُرِيد', 'مَنصُور', 'عَادِل', 'فَرَح', 'ارْشَد',  'عُلَمَاء', 'سُلطَنَة', 'خِلَافَة']
features = {
    'pos': 'noun'
}

for i in range(len(lemmas)):
  print(lemmas[i])
  analyses = generator.generate(lemmas[i], features)

  # Extract and print unique diacritizations from generated analyses
  for diac in set([a['diac'] for a in analyses]):
      print(diac)

```

```python
from camel_tools.tokenizers.word import simple_word_tokenize
from camel_tools.disambig.mle import MLEDisambiguator

mle = MLEDisambiguator.pretrained()

# The disambiguator expects pre-tokenized text
sentence = simple_word_tokenize('بسم الله الرحمان الرحيم.\n و صلى الله على سيدنا محمد نبيه و اله و صحبه و سلم')

disambig = mle.disambiguate(sentence)

# For each disambiguated word d in disambig, d.analyses is a list of analyses
# sorted from most likely to least likely. Therefore, d.analyses[0] would
# be the most likely analysis for a given word. Below we extract different
# features from the top analysis of each disambiguated word into seperate lists.
diacritized = [d.analyses[0].analysis['diac'] for d in disambig]
pos_tags = [d.analyses[0].analysis['pos'] for d in disambig]
lemmas = [d.analyses[0].analysis['lex'] for d in disambig]

# Print the combined feature values extracted above
for triplet in zip(diacritized, pos_tags, lemmas):
    print(triplet)
```
