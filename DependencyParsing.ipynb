```python
import stanza
import spacy_stanza
import spacy
from spacy import displacy
```

```python

processor_dict = {
    'depparse': 'default',
    'tokenize': 'default', 
    'pos': 'default', 
    'lemma': 'default',
    'ner':'default',
    'mwt':'padt'
}
# download the arabic model with dependency parsing and create a pipeline
stanza.download('ar', processors=processor_dict)

# tokenization: https://stanfordnlp.github.io/stanza/tokenize.html#tokenization-and-sentence-segmentation
# sometimes nlp here does not split sentences based on their punctuation, 
# so by setting tokenize_pretokenized=True here, I control how the sentences are tokenized: as newline (\n) separated sentences.
nlp = spacy_stanza.load_pipeline('xx', lang='ar', tokenize_pretokenized=True)
#nlp = spacy_stanza.load_pipeline('xx', lang='ar')
```

```python
doc = nlp('بسم الله الرحمان الرحيم.\n و صلى الله على سيدنا محمد نبيه و اله و صحبه و سلم')

for token in doc:
  print(token.text, token.lemma_, token.pos_, token.dep_, token.ent_type_)
#print(doc.ents)

for i, sentence in enumerate(doc.sents):
  print(sentence.)
```

```python
displacy.render(doc, style="dep",jupyter=True)
```

```python
stanzanlp = stanza.Pipeline(lang='ar', processors='tokenize,mwt,pos,lemma,depparse')
doc = stanzanlp('بسم الله الرحمان الرحيم.\n و صلى الله على سيدنا محمد نبيه و اله و صحبه و سلم')
print(*[f'id: {word.id}\t\tword: {word.text}\t\thead id: {word.head}\t\thead: {sent.words[word.head-1].text if word.head > 0 else "root"}\t\tdeprel: {word.deprel}' for sent in doc.sentences for word in sent.words], sep='\n')
```
